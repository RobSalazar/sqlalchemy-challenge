# Bonus: Temperature Analysis I

```python
import pandas as pd
from datetime import datetime as dt
from scipy import stats
```

```python
# "tobs" is "temperature observations"
df = pd.read_csv('resources/hawaii_measurements.csv')
df.head()
```

```python
# Set the date column as the DataFrame index
# Convert the date column format from string to datetime
date_df = df.set_index("date")
date_df.index = pd.to_datetime(date_df.index)
date_df
```

```python
# Drop the prcp column
clean_df = date_df.reset_index()

df = clean_df.drop(columns = ['prcp'])
df.head()
```

### Compare June and December data across all years 

```python
# Filter data for desired months
june_df = df[df['date'].dt.month == 6]
june_df
```

```python
#Looking at the standard dev. for june data
june_df.describe()
```

```python
# Filter data for desired months
december_df = df[df['date'].dt.month == 12]
december_df
```

```python
#Looking at the standard dev. for december data
december_df.describe()
```

```python
# Identify the average temperature for June
june_avg = june_df['tobs'].mean()
june_avg
```

```python
# Identify the average temperature for December
dec_avg = december_df['tobs'].mean()
dec_avg
```

```python
# Run paired t-test
stats_values = stats.ttest_ind(june_df["tobs"], december_df["tobs"])
stats_values
```

### Analysis

In order to run a ttest_rel (related) we would need to have the same length of data for both a and b samples. This analysis is very telling for the temperature in Hawaii. What this data tells me with the averages being close and the p value rejecting the null hypothesis is that the temperature doesnt really change. This is extremely reinforced with the dataframes above and the t test.
